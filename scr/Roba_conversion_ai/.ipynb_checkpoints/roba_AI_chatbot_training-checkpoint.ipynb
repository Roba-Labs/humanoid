```python
from chatterbot import ChatBot
import chatterbot
from chatterbot.trainers import ChatterBotCorpusTrainer ,ListTrainer
from chatterbot import filters
import json
import io 
import numpy as np
import glob
import sys
sys.path.insert(1, '/home/roba/Robot_roba/scr/Roba_conversion_ai')  
sys.path.insert(1, '/home/roba/Robot_roba/scr/Roba_conversion_ai/data')  
QAA=[]

for file in glob.glob("data/*.json"):
    with open(file, 'r') as stream:
        data_loaded = json.loads(stream.read())
        
        for x, data in enumerate(data_loaded):
               
              q = np.array(data['question'])
              a = np.array(data['answer']) 


              if q.size>1:
                   for qq in q:
                       if a.size>1:
                           for aa in a:
                               QAA.append([str(qq),str(aa)])
                       else:
                             QAA.append([str(qq),str(a)])

              else:
                if a.size>1:
                   for aa in a:
                       QAA.append([str(q),str(aa)])
                else:
                     QAA.append([str(q),str(a)])

                    
                    

robabot = ChatBot('robabot', read_only=True,
   
    storage_adapter={
        'import_path': 'chatterbot.storage.SQLStorageAdapter',
        'database_uri': 'sqlite:///data/robadata.db'
    },
    
    logic_adapters=[
        'chatterbot.logic.MathematicalEvaluation',
    {
         
        'import_path':'chatterbot.logic.roba_ai.BestMatch',
        'default_response': 'I am sorry, but I do not understand.'
    }
    
    ]    
    
)

QAA = {'conversations': QAA}
with open('data/data1/dataroba.json', 'w') as f:
    json.dump(QAA, f)
                    
trainer = ChatterBotCorpusTrainer(robabot)
trainer.train('data/data1/dataroba.json')
print("finished")
```
